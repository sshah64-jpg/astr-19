```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.uniform(0, 1, 1000)

plt.hist(data, bins=100, color='red', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Uniform Random Number Distribution')
```

```python
plt.savefig('uniform_histogram.pdf')

plt.show()
```
