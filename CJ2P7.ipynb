```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
```

```python
def exp(x):
    return np.exp(x)

y = exp(x)
```

```python
plt.figure
plt.plot(x, y)

plt.xlabel("Time in Milliseconds")
plt.ylabel("Awesomeness")

plt.savefig("Milliseconds_Awesomeness.pdf")

plt.show
```
