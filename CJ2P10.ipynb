```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=100, color='teal', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normally Distribution of Random Numbers')
```

```python
plt.savefig('normal_distribution_histogram.pdf')

plt.show()
```
